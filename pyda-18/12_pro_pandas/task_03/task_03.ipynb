### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas
import numpy

def fetch_ratings() -> pandas.DataFrame:
    return pandas.read_csv('ratings.csv')

def flat_map_to_mean_time_row(row: pandas.DataFrame) -> pandas.Series:
    '''
    Finds the difference between the maximum and minimum value of a timestamp column
    '''
    time_life = numpy.nan
    if len(row) > 100:
        time_life = row['timestamp'].max() - row['timestamp'].min()

    return pandas.Series(
        data={'timeLife': time_life},
        index=['timeLife']
    )

def get_life_time_table(ratings: pandas.DataFrame) -> pandas.DataFrame:
    '''
    Groups the ratings table by user and finds the lifetime
    '''
    
    filter_ratings = ratings.loc[:, ['userId', 'timestamp']]
    life_time_table = filter_ratings.groupby('userId').apply(flat_map_to_mean_time_row).reset_index()
    clear_life_time_table = life_time_table.loc[~pandas.isnull(life_time_table['timeLife']), :]
    
    return clear_life_time_table

ratings = get_life_time_table(
    ratings=fetch_ratings()
) 
ratings.head(20)
```
