### Задание 1:

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

### Решение

```python
import numpy
from matplotlib import pyplot
from scipy import stats
```

```python
# Нормальное

def show_normal_graphs():
    for index in range(5):
        mu = numpy.random.uniform(1, 3, 1)[0]
        sigma = numpy.random.uniform(0, 1, 1)[0]
        
        norm_rv = stats.norm(
            loc=mu,
            scale=sigma)
        norm_rv.rvs(size=10)
        
        x_values = numpy.linspace(0, 4, num=100)

        pdf = norm_rv.pdf(x_values)
        pyplot.plot(x_values, pdf)
        pyplot.ylabel('$F(x)$')
        pyplot.xlabel('$x$')
    pyplot.show()
    
show_normal_graphs()
```

```python
# Экспоненциальное

def show_exponential_graphs():
    for index in range(5):
        lam = numpy.random.uniform(0.1, 1, 1)[0]
        expon_rv = stats.expon(lam)
        expon_rv.rvs(size=10)

        x_values = numpy.linspace(0, 4, num=1000)
        pdf = expon_rv.pdf(x_values)
        pyplot.plot(x_values, pdf)
        pyplot.ylabel('$F(x)$')
        pyplot.xlabel('$x$')
    pyplot.show()
    
show_exponential_graphs()
```

```python
# Стьюдента

def show_student_graphs():
    for index in range(5):
        n = numpy.random.uniform(0.1, 2, 1)[0]
        student_rv = stats.t(n)
        x_values = numpy.linspace(-4, 4, num=1000)

        student_rv.rvs(size=10)
        
        pdf = student_rv.pdf(x_values)
        pyplot.plot(x_values, pdf)
        pyplot.ylabel('$F(x)$')
        pyplot.xlabel('$x$')
    pyplot.show()
    
show_student_graphs()
```

### Задание 2

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания  в отрезок $[3.5;7]$, т.е. P(3.5 &lt; $\xi$ &lt; 7).

### Решение

```python
import numpy
from matplotlib import pyplot
from scipy import stats

def show_continuous_graphs():
    # properties
    func_a, func_b = 0, 5
    segment_a, segment_b = 3.5, 7
    
    # formulas
    C = 1 / (func_b - func_a)
    math_expectation = (func_a + func_b) / 2
    probability_hitting_segment = (func_b - segment_a) * C
    
    # data graph
    uniform_rv = stats.uniform(func_a, func_b)
    uniform_rv.rvs(size=10)
    x_values = numpy.linspace(-1, 8, num=1000)
    pdf = uniform_rv.pdf(x_values)
    
    # show graph
    pyplot.plot(x_values, pdf)
    pyplot.ylabel('$F(x)$')
    pyplot.xlabel('$x$')
    pyplot.axvline(
        x=segment_a, 
        color='gray',
        linestyle='--')
    pyplot.axvline(
        x=segment_b, 
        color='gray',
        linestyle='--')
    pyplot.show()
    
    # prints
    print('C =', C)
    print('Математическое ожидание =', math_expectation)
    print('Вероятность попадания в отрезок  {} <  𝜉  < {} = {:.2f}'.format(segment_a, segment_b, (func_b - segment_a) * C))
    
show_continuous_graphs()
```

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

### Решение

```python
import numpy

def print_bits(red_sec, yellow_sec, green_sec):
    total = red_sec + yellow_sec + green_sec
    
    red_probability = red_sec / total
    yellow_probability = yellow_sec / total
    green_probability = green_sec / total

    red_bit = red_probability * -numpy.log2(red_probability)
    yellow_bit = yellow_probability * -numpy.log2(yellow_probability)
    green_bit = green_probability * -numpy.log2(green_probability)

    print('Сообщение о цвете светофора в текущий момент:')
    print('\t красный: {:.3} бит'.format(red_bit))
    print('\t желтый: {:.3} бит'.format(yellow_bit))
    print('\t зеленый: {:.3} бит'.format(green_bit))
    
print('Светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным')
print_bits(
    red_sec=54,
    yellow_sec=4,
    green_sec=28)

print('\nСветофор горит 28 секунды красным, 28 секунды жёлтым и 28 секунд зелёным')
print_bits(
    red_sec=28,
    yellow_sec=28,
    green_sec=28)
```
