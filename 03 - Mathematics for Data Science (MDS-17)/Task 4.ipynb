## Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

## Решение

Первая производная для $f(x) = x^5 + 4sin(2x) + cos(3x+3)$:

$f'(x) = 5x^4 + 8cos(2x) - 3sin(3x+3)$

$f'(1) = 5 * 1^4 + 8cos(2 * 1) - 3sin(3 * 1 + 3) = 5 + 8cos(2) - 3sin(6)$

---

Вторая производная $f'(x) = 5x^4 + 8cos(2x) - 3sin(3x+3)$:

$f''(x) = 20x^3 - 16sin(2x) - 9cos(3x+3)$

$f''(1) = 20x^3 - 16sin(2x) - 9cos(3x+3) = 20 * 1^3 - 16sin(2 * 1) - 9cos(3 * 1 + 3) = 20 - 16sin(2) - 9cos(6)$

## Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

## Решение

$f(x) = sin^5(2x+1)$

$f'(x) = 10cos^4(2x+1)$

$f''(x) = -80sin^3(2x+1)$

```
import numpy
import matplotlib.pyplot as pyplot


def show_graph(func_x, axis_y, title):
    pyplot.plot(axis_y, func_x)
    pyplot.title = title
    pyplot.show()

def first_derivative(x):
    return 10 * numpy.cos(2 * x + 1) ** 4

def second_derivative(x):
    return -80 * numpy.sin(2 * x + 1) ** 3

axis_y = numpy.linspace(-5, 5, 1000)


show_graph(
    func_x=first_derivative(axis_y),
    axis_y=axis_y,
    title='10cos^4(2x+1)')

show_graph(
    func_x=second_derivative(axis_y),
    axis_y=axis_y,
    title='-80sin^3(2x+1)')
```
